#### Import all the libraries that are necessary.

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# 1st way (call `subplots` twice using the `index` parameter) 
# subplot(nrows, ncols, index, **kwargs)
plt.subplot(2, 1, 1)
plt.plot(x, y)
# subplot(pos, **kwargs)
plt.subplot(212)
plt.scatter(x, z)
```

```python
# 2st way (call `subplots` only once not using the `index` parameter) 
# f, (ax1, ax2) 
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax2.scatter(x, z)
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# subplot(nrows, ncols, index, **kwargs)
fig = plt.figure(figsize=(8,5))
plt.subplot(2, 1, 1)
plt.plot(x, y)
# subplot(pos, **kwargs)
plt.subplot(212)
plt.scatter(x, z)
```

```python
# f, (ax1, ax2) 
fig = plt.figure(figsize=(4,3))
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax2.scatter(x, z)
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# subplot(nrows, ncols, index, **kwargs)
fig = plt.figure(figsize=(8,5))
plt.subplot(2, 1, 1)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.plot(x, y)

# subplot(pos, **kwargs)
plt.subplot(212)
plt.xlabel('xlabel')
plt.ylabel('zlabel')
plt.scatter(x, z)
```

```python
# f, (ax1, ax2) 
fig = plt.figure(figsize=(4,3))
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax1.set_xlabel(xlabel='x')
ax1.set_ylabel(ylabel='y')
ax2.scatter(x, z)
ax2.set_xlabel(xlabel='x')
ax2.set_ylabel(ylabel='z')
```
