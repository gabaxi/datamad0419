#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users_columns = pd.DataFrame(users_table,columns=['userId', 'Reputation','Views','UpVotes','DownVotes'])
users_columns.head(5)
```

```python
posts_columns = pd.DataFrame(posts_table,columns=['postId', 'Score','userId','ViewCount','CommentCount'])
posts_columns.head(5)
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
# DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, 
# sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None
#new_df = users_columns.merge(posts_columns, how='left', left_on='userId', right_on='userId')
new_df = pd.merge(users_columns, posts_columns, on='userId')
new_df.head(10)
```
